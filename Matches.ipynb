# **Data Visulization**

```python
sns.countplot(x='season',data =df,color='green')
plt.title("Seasons played by Indian Premiem League")
```

```python
sns.countplot(x='winner',data=df,color="yellow")
import matplotlib.pyplot as plt
plt.xticks(rotation=90)
plt.title("Winning Team")
```

```python
sns.countplot(x='venue',data=df,order=df['venue'].value_counts().index , color="purple")
plt.xticks(rotation=90)
plt.title("Indain Premian Leaguge Venue")
```

```python
sns.barplot(x=df['player_of_match'].value_counts()[:10].index, y=df['player_of_match'].value_counts()[:10],color='orange')
plt.xticks(rotation=90)
plt.title("Player of match")
```
